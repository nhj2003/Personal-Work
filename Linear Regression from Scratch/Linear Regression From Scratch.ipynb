```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
import random
from sklearn import datasets
```

```python
# Basic data to use with regression created from scratch
X, y = datasets.make_regression(
    n_samples = 500, n_features = 1, noise = 15)
```

```python
sns.scatterplot(x = X.reshape(-1), y = y)
```

```python
class LinearRegression:
    def __init__(self, lr = 0.01, n_iters = 1000):
        self.lr = lr
        self.n_iters = n_iters
        self.weights = None
        self.bias = None

    #Fit method within Linear Regression
    def fit(self, X, y):
        num_samples, num_features = X.shape
        # Weights are randomly generated, so that we can use gradient descent to reach the 
        # optimal. If we set them at a certain value, it could be that we get stuck in a "valley"
        # more often than if we just use random values
        self.weights = np.random.rand(num_features)
        self.bias = 0

        # Undergo gradient descent for number of specified iterations
        for i in range(self.n_iters):
            # Get predictions
            y_pred = np.dot(X, self.weights) + self.bias

            # Partial derivative calculations
            dw = (1 / num_samples) * np.dot(X.T, y_pred - y)
            db = (1 / num_samples) * np.sum(y_pred - y)

            self.weights = self.weights - self.lr * dw
            self.bias = self.bias - self.lr * db

        return self

    def predict(self, X):
        return np.dot(X, self.weights) + self.bias
```

We use matrix multiplication to end up with the final predictions.
X is in a matrix of N x M, (num_rows, num_features)
Weights are in their own matrix of N x 1, (num_features, ) 
If we then matrix multiply the two together, we can multiply the weights for each feature with its respective "feature spot" in each row in X. This can be accomplished with the numpy function: np.dot(X, weights). The weights initialized are not going to be correct, so we use gradient descent to slowly optimize the weights to one that will minimize the error.

The loss function is as follows: 
<br> $MSE = J(w, b) = \frac{1}{N}\sum^n_{i = 1}{(y_i - (wx_i + b))^2}$
<br> The $wx_i + b$ represents the predictions we get by multiplying the weights $w$ and the specific $x_i$ value, plus any bias $b$ that would be added on to ensure there is some offset for when data doesn't go through the origin. Then we subtract each specific predicted $wx_i + b$ value by its real $y_i$ value and sum the square of the resulting residuals. From there we divide by the total number of rows to get the mean squared error.

Gradient Descent:
We want to minimize the loss function, so we find need to find where the gradient (the derivative of the loss function) ends up being 0. Since the gradient always points in the direction where the function is increasing the fastest, we need to go in the opposite direction of the gradient when we compute it. There are two "variables" we can manipulate, so we take the partial derivative with respect to both the weights and the bias values. Thus, we end up with
$$ J'(m,b) = 
\begin{bmatrix}
\frac{\partial J}{\partial w} \\
\frac{\partial J}{\partial b}
\end{bmatrix} =
\begin{bmatrix}
\frac{1}{N} \sum -2x_i \left( y_i - (wx_i + b) \right) \\
\frac{1}{N} \sum -2 \left( y_i - (wx_i + b) \right)
\end{bmatrix} $$
<br> Once we calculate the derivatives using the above equation, we need to then accordingly adjust the weights and bias. With $\alpha$ as your learning rate (which adjusts how fast you want to "optimize" the equation), we update as follows:
<br> $$ w = w - a\cdot dw $$ $$b = b - a\cdot db$$
We subtract against $a\cdot dw$ and $a\cdot db$ because we want to go in the opposite direction of the gradient, to go in the direction where the function is most decreasing to minimize the error. We simplify the derivatives by turning $wx_i + b$ into $\hat{y}$ to end up with the resulting equations:
$$ \begin{bmatrix}
\frac{\partial J}{\partial w} \\
\frac{\partial J}{\partial b}
\end{bmatrix} = 
\begin{bmatrix}
\frac{1}{N} \sum^n_{i = 1} 2x_i(\hat{y}-y_i) \\
\frac{1}{N} \sum^n_{i = 1} 2(\hat{y} - y_i)
\end{bmatrix} $$
Where $N$ = len(rows) and $x_i(\hat{y} - y_i)$ is calculated by X transposed multiplied by y_pred - y. We then do this iteratively by however many times we specified it in the code to get our final model.

Prediction is simply just matrix multiplying X by the weights plus whatever bias we ended with to get the final value.

```python
model = LinearRegression()
model.fit(X = X, y = y)
```

```python
exes = [X.min(), X.max()]
whys = model.predict(np.array(exes).reshape(-1, 1))
sns.scatterplot(x = X.reshape(-1), y = y)
sns.lineplot(x = exes, y = whys, color = 'orange', linewidth = 3)
```
