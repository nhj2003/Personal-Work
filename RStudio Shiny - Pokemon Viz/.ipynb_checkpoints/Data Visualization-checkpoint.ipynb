```python
poke = pd.read_csv('Pokemon.csv')
poke.head()
```

```python
stats = poke.groupby(by = 'Generation').describe()
```

```python
print('\033[1m Total \033[0m \n', stats['Total'])
print('\033[1m HP \033[0m \n', stats['HP'])
print('\033[1m Attack \033[0m \n', stats['Attack'])
print('\033[1m Defense \033[0m \n', stats['Defense'])
print('\033[1m Sp. Atk \033[0m \n', stats['Sp. Atk'])
print('\033[1m Sp. Def \033[0m \n', stats['Sp. Def'])
print('\033[1m Speed \033[0m \n', stats['Speed'])
```
